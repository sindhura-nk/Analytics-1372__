## Scenario2: Add a column to the dataframe that displays Price tags
    Price <15 : Cheap
    Price >15 and <25: Affordable
    Price >25: Expensive

```python
car["Price"][0]
```

```python
price_tags = []
for i in car["Price"]:
    if i>0 and i<15:
        price_tags.append("Cheap")
    elif 15<=i<25:
        price_tags.append("Affordable")
    elif i>25:
        price_tags.append("Expensive")
    else:
        price_tags.append("Invalid")     
print(price_tags)
```

```python
len(price_tags)
```

```python
car.shape
```

```python
car["Price_Tags"] = price_tags 
```

## Joins
Joins are used to combine data from mutliple tables.

```python
emp = {
    "EmpId":[1,2,3,4,5],
    "Emp_Names":["Raman","Thaman","Suman","Baman","Chaman"],
    "Salary":[56000,45000,67000,89000,10000],
    "deptID":[101,103,107,108,114]
}
```

```python
dept = {
    "dept_ID":[101,102,103,104,105,106,107,108],
    "Dept_names":["Finance","HR","Sales","IT","BPO","KPO","Admin","Marketting"]
}
```

```python
emp_df = pd.DataFrame(emp)
dept_df = pd.DataFrame(dept)
```

```python
emp_df
```

```python
dept_df
```

## Retrieve employee details working in respective departments

```python
pd.merge(left=emp_df,right=dept_df,how="inner",left_on="deptID",right_on="dept_ID")
```
