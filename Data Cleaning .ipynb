## Handling Missing Data
    If the missing data of the respective feature is minimum(<10% data), you can drop the data(rows that contain missing data)
    syntax: df = df.dropna(subset=[col_name]) # by default axis=0- row wise dropping of data
    If the missing data of the respective feature is more(<50% data), then you can replace the missing data with Central Tendency concepts
    If the missing data of the respective feature is more than 50% data,
    1. Is the feature important for the analysis, you have to handle it
    2. Is the feature not important, drop the feature(column) 
    : syntax: df = df.dropna(subset=[col_name],axis=1)
    : syntax: df= df.drop(columns=[col_name])

```python
def handle_outlier(col):
    global cars_new
    # 1st quartile - 25%
    q1 = np.percentile(cars_new[col],25)
    # 3rd quartile - 75%
    q3 = np.percentile(cars_new[col],75)
    # IQR - Inter quartile range
    iqr = q3-q1
    # calculate minimum and maximum to determine outliers
    mini = q1 - (1.5*iqr)
    maxi = q3 + (1.5*iqr)

    lower_index = list(cars_new[cars_new[col]<mini].index)
    upper_index = list(cars_new[cars_new[col]>maxi].index)

    all_ = lower_index + upper_index
    cars_new = cars_new.drop(index=all_)
    print("====================================================")
    print(col)
    print(f"Shape before outlier handling is {cars_new.shape}")
    print(f"Minimum values present at index {lower_index}")
    print(f"Maximum values present at index {upper_index}")
    print(f"Outliers are handled for {col}")
    print(f"Shape after outlier handling is {cars_new.shape}")
    print("===============================")
    #return cars_new
```

```python
con = list(cars_new.columns[cars_new.dtypes !="object"])
```

```python
con
```

```python
for i in con:
    handle_outlier(i)
```

## Data Cleaning using pipelines

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
cars_new.isna().sum()
```

```python
con
```

```python
cat = list(cars_new.columns[cars_new.dtypes=="object"])
cat
```

```python
num_pipe = make_pipeline(SimpleImputer(strategy="mean")) # u can also provide median
```

```python
cat_pipe = make_pipeline(SimpleImputer(strategy="most_frequent")) # we place categorical data with mode. Term to be passed
# for mode is most_frequent
```

```python
pipe = ColumnTransformer([
    ('categorical',cat_pipe,cat), # handles missing data for categorical columns
    ('continouous',num_pipe,con) # handles missing data for continuous columns
]).set_output(transform='pandas') 
# setoutput method converts the array output into dataframe
```

```python
pipe
```

```python
cars_new_cleaned = pipe.fit_transform(cars_new)
```
