## Query method is also used for filtering of content
### Syntax: dataframe_name.query("col_name conditional statement")
    1. Always pass the content in double quotes
    2. column name must NOT be included in quotes
    3. provide condition and pass the text content in single quotes
    example: car.query("Type=='Compact'")
    car.query('Type=="Compact"')

```python
# find out cars which belong to Compact type
car.query("Type=='Compact'")
```

### Sort the data
### sort_values() method: ascending order, descending order
    categorical/text : Asc - A-Z (a-z). Desc- Z-A (z-a)
    Numerical : Asc: smallest to largest , Desc: Largest to smallest
### Syntax: dataframe_name.sort_values(by=[col_name])
default: ascending order
### Syntax: dataframe_name.sort_values(by=[col_name],ascending=False)

## To calculate aggregated details, you can also use agg()
    df.agg({col_name:[agg_func]})
    df.agg({col_name:["sum","min"],col_name2:["sum","min"]})

```python
car.agg({"Price":["min","max","sum"]})
```

```python
car.agg({"Price":["min","max","sum"],
         "Weight":["min","max","sum"]})
```

```python
car.agg({"Price":["mean","std"],
        "Horsepower":["mean","std"]}).round(2)
```

## Groupby
## syntax: df.groupby(by=[col_name,col_name2]).aggfunc()[[col_subsetting]]

## scenario : Find out min prices for all types of cars

```python
car.groupby(by=["Type"]).min(numeric_only=True)
```

```python
car.groupby(by=["Type"]).min(numeric_only=True)[["Price"]]
```

## Show the total price of cars on basis of airbags

```python
car.groupby(by=["AirBags"]).sum(numeric_only=True)
```

```python
car.groupby(by=["AirBags"]).sum(numeric_only=True)[["Price"]]
```
