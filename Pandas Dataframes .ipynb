## Dataframe Subsetting
    loc
    syntax: dataframe_name.loc[row_name,col_name]
    examples: df.loc[1:3,col_name1:col_name4] # 1 and 3 are included in output, col_name1 and col_name4 are included in output
    df.loc[[1,2],[col_name1,col_name3]]
    
    iloc
    syntax: dataframe_name.iloc[row_index,col_index]
    # default start index is 0
    # default stop index is len of the elements
    examples: df.loc[1:3,col_name1:col_name4] # 1 ,until  2 are retrieved in output, col_name1,until col_name3 are included in output
    df.iloc[[1,2],[1,3]]

## Filtering
    dataframe_name[col_name] provide the condition
    
    filtering condition = dataframe_name[col_name]>100000 
    this will return boolean output(True and False)
    
    To get the output, 
    dataframe_name[filtering condition]
    dataframe_name[dataframe_name[col_name]>100000]

```python
# filtering condition
fil1 = df2["PROFIT"]>100000
df2[fil1]
```
